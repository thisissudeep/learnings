<h1 style='color:red; text-align:center'>Using Scikitlearn OHE</h1>


<h1 style='color:red'>Raw Data</h1>


```python
import pandas as pd

df = pd.read_csv("carprices.csv")
df
```

<h1 style='color:red'> Labeled Encoded column</h1>


```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["Car Model"] = le.fit_transform(df["Car Model"])
df
```

<h1 style='color:red'>Onehotencoded columns (x and y)</h1>


```python
x = df[["Car Model", "Mileage", "Age(yrs)"]].values
y = df["Sell Price($)"].values
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer([("Car Model", OneHotEncoder(), [0])], remainder="passthrough")
x = ct.fit_transform(x)
print(x, "\n")
x = x[:, 1:]
print(x, "\n")
print(y)
```

<h1 style='color:red'>Model</h1>


```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x, y)
print(model.predict([[0, 1, 45000, 4]]))
print(model.predict([[1, 0, 86000, 7]]))
print(model.score(x, y))
```
