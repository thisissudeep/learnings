**How to draw those straight lines?:**

You might remember about **slope-intercept equation** to draw a straight line from your high school days math class. Home prices can be presented as following equation,

y = m * x + b

home price = ( slope * area ) + Y intercept

<h2>STEPS TO BUILD THE MODEL:</h2>

1) **Scatter plot** the given values

2) Draw **Multiple Lines** using slope-intercept equation

3) Take **1 1ine** where sum of total error is minimum

your model is now ready for prediction

```python
import pandas as pd
import numpy as np
from sklearn import linear_model   #used for linear regression
import matplotlib.pyplot as plt    #used for Data visualization
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
%matplotlib inline           
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,df.price)
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US $)')
```

```python
new_df = df.drop('price',axis='columns')
new_df
```

```python
price = df.price
price
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,price)
```

**(1) Predict price of a home with area = 3300 sqr ft**

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

**Y = m * X + b (m is coefficient and b is intercept)**

```python
3300*135.78767123 + 180616.43835616432
```

**(2) Predict price of a home with area = 5000 sqr ft**

```python
reg.predict([[5000]])
```

<h3 style="color:purple">Generate CSV file with list of home price predictions</h3>

```python
area_df = pd.read_csv("areas.csv")
area_df.head(3)
```

```python
p = reg.predict(area_df)
p
```

```python
area_df['prices']=p
area_df
```

```python
area_df.to_csv("prediction.csv")
```
