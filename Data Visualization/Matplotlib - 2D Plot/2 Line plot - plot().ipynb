<h1 align="center" style="color:green">plot() function</h1>

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

<h3 style="color:purple">Format string (Color,marker,linestyle parameters)</h3>

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y,'b+--')
```

```python
# Same effect as 'b+' format string
plt.plot(x,y,color='blue',marker='+',linestyle='')  # here it has blank linestyle
```

<h3 style="color:purple">markersize parameter</h3>

```python
plt.plot(x,y,color='blue',marker='+',linestyle='',markersize=20)
```

<h3 style="color:purple">alpha parameter to control transparency of line chart</h3>

```python
plt.plot(x,y,'g<',alpha=0.5) # alpha can be specified on a scale 0 to 1
```
