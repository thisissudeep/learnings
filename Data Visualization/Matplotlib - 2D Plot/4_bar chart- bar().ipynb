<h1 align="center", style="color:green">Matplotlib Tutorial: Bar Chart</h1>

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

<h3>Simple bar chart showing revenues of major US tech companies</h3>

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos = np.arange(len(company))  #a numpyarray(2d array) , we can use list(1d array) also but when performing -0.2,=0.2 array would perform better
xpos
```

```python
plt.xticks(xpos,company)  #for inserting company names in xpos
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')

plt.bar(xpos,revenue, label="Revenue")
plt.legend()
```

<h3>Multiple Bars showing revenue and profit of major US tech companies</h3>

```python
profit=[40,2,34,12]
```

```python
plt.bar(xpos,revenue, label="Revenue")
plt.bar(xpos,profit,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")    #0.2 here indicating the position
plt.bar(xpos+0.2,profit, width=0.4,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

<h3>Horizontal bar chart using barh function   (h-horizontal)<h3>

```python
plt.barh(xpos,revenue, label="Revenue")

plt.yticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
```
