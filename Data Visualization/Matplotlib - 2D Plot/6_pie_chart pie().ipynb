<h1 align="center", style="color:green">Matplotlib Tutorial: Pie Charts</h1>

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

<h3 style="color:blue">Draw pie chart to track down home expenses</h3>

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals)
```

<h3 style="color:purple">labels parameter & Draw pie chart as perfect circle</h3>

plt.show() - to eliminate text in output

plt.axis() - to set the axis Range of the current plot. It allows you to specify the minimum and maximum values ( Range)for the x-axis and y-axis


```python
plt.pie(exp_vals,labels=exp_labels)
plt.axis("equal")
plt.show()
```

<h3 style="color:purple">shadow, autopct, radius </h3>

shadow - provides shadow 

autopct- gives percentage to labels [0.2 means 47.30 indicating the points length ]

radius - default is 1 size. radius =1.5 means increasing size of pie chart

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%0.2f%%',radius=1.5)
plt.show()
```

<h3 style="color:purple">Explode</h3>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.4]) # here 0's in explode represents the no. of labels
plt.show()
```


<h3 style="color:purple">counterclock and startangle</h3>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2],counterclock=True, startangle=45)
plt.show()
```
